```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
marketing_customer_analysis = pd.read_csv('marketing_customer_analysis.csv')
type(marketing_customer_analysis)
```

```python
marketing_customer_analysis.head()
marketing_customer_analysis.columns
```

```python
# the dataframe is a mixture of numerical and categorical data
```

```python
## Show a plot of the total number of responses.

response_counts = marketing_customer_analysis.Response.value_counts()
#type(marketing_customer_analysis.Response.value_counts())
print(response_counts)

No = response_counts[0]
Yes = response_counts[1]
```

```python
type(response_counts)
```

```python
fig,ax = plt.subplots(figsize=(8,6))
ax = plt.bar(["No","Yes"],response_counts)
plt.show()
```

### Show a plot of the response rate by the sales channel.

```python
sns.countplot(x="Sales Channel", hue="Response", data=marketing_customer_analysis)
plt.show()
```

### Show a plot of the response rate by the total claim amount.

```python
# Check vertical axis

sns.countplot(x="Total Claim Amount", hue="Response", data=marketing_customer_analysis, linewidth=5
)
plt.show()
```

### Show a plot of the response rate by income.

```python
# Check vertical axis

sns.countplot(x="Income", hue="Response", data=marketing_customer_analysis)
plt.show()
```
